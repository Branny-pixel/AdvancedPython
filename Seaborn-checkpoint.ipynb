```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_excel("cars.xlsx")
car.head()
```

```python
variables=["Min Price (Lakh)","Max Price (Lakh)","Range (kmpl)","CC","Seats","Variants","Ex-Showroom Price","RTO","Insurance","Other","Onroad Price"]
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables,start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=car[var],y=car['Min Price (Lakh)'],alpha =0.7)
    plt.title(f'scatter plot of {var}) vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show
```

```python
#histogram for all independent variables
plt.figure(figsize=(15,10))
for i,var in enumerate(variables,start=1):
    plt.subplot(3,4,i)
    sns.histplot(car[var],bins=20, kde=True, color='green', alpha=1)
    plt.title(f'scatter plot of {var})')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show
```

```python
sns.boxplot(data=car,x="Type",y="Range (kmpl)",hue="Type",palette=["Brown","Red"])
plt.title("Boxplot of Range")
plt.xlabel("Type")
plt.ylabel('Range (kmpl)')
plt.show()
```

```python
selected_columns=["Min Price (Lakh)","Max Price (Lakh)","Range (kmpl)","CC","Seats","Variants","Ex-Showroom Price","RTO","Insurance","Other"]
df_selected = car[selected_columns]
df_selected = car[selected_columns]
df_selected.head()
```

```python
#Correlation
corr=df_selected.corr()
corr
```

```python
sns.heatmap(corr, annot=True,cmap="Blues")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
#Violin plot of max price by car Type
sns.violinplot(data=car,x="Type",y="Max Price (Lakh)", hue="Type",palette=["Orange","Red"])
plt.title("Violin Plot of Maximum Price by Car Type")
plt.show()
```

```python
sns.violinplot(data=car,x="Type",y="Range (kmpl)", hue="Type",palette=["Blue","Green"])
plt.title("Violin Plot of Range by Car Type")
plt.show()
```
